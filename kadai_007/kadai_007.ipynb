```python
# リストの作成
weekly_weather_list = [
    "月曜日は晴れです",
    "火曜日は雨です",
    "水曜日は晴れです",
    "木曜日は晴れです",
    "金曜日は曇りです",
    "土曜日は曇りのち雨です",
    "日曜日は雷雨です"
]

# データの取り出し例（0番目＝月曜日）
# リスト：インデックス2（3番目）の要素を取得
print(weekly_weather_list[2]) 
# 出力結果: 水曜日は晴れです
```
