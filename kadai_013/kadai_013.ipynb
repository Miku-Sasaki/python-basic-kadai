```python
# 消費税を加えた計算結果を返す関数
def calculate_total_price(price, tax_rate):
    total = price * (1 + tax_rate / 100)
    return total

# このファイルが「直接実行された場合」のみ、以下のブロックが動く
if __name__ == "__main__":
    # 引数の設定
    price_input = 110
    tax_input = 10
    
    # 関数の呼び出し
    result = calculate_total_price(price_input, tax_input)
    
    # 結果の表示
    print(f"商品の金額: {price_input}円")
    print(f"消費税率: {tax_input}%")
    print(f"合計金額（税込）: {result}円")
```
