```python
# Humanクラスの定義
class Human:
    def __init__(self, name, age):
        # 属性（プロパティ）の設定
        self.name = name
        self.age = age

    # 情報を出力するメソッド
    def printinfo(self):
        print(f"名前: {self.name}")
        print(f"年齢: {self.age}")

# --- プログラム内での使用 ---

# インスタンスを作成し、変数「person」に代入
person = Human("侍 太郎", 25)

# メソッドを呼び出して情報を表示
person.printinfo()
```
